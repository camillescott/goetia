```python
import pandas as pd
import seaborn as sns
from ficus import FigureManager
from IPython.display import FileLink
```

```python
meta_cols = ['full', 'tip', 'island', 'trivial']
data = pd.read_csv('mmetsp/Asterionellopsis_glacialis/Ast_g.unode.meta.csv')
del data['unknown']
```

```python
with FigureManager(filename='Ast_g.n_unode_meta.png', ncols=2, figsize=(16,8), show=True) as (fig, ax_mat):
    data[meta_cols].plot(ax=ax_mat[0], title='Number of Unitig-Node Types', lw=4)
    ax_mat[0].set_xlabel('Read Number')
    
    data[meta_cols].divide(data['n_unodes'], axis=0).plot(ax=ax_mat[1], lw=4,
                                                          title='Proportion of Unitig-Node Types')
    ax_mat[1].set_xlabel('Read Number')
FileLink('Ast_g.n_unode_meta.png')
```

```python
with FigureManager(filename='Ast_g.n_tags.png', figsize=(12,12), show=True) as (fig, ax):
    data[['n_dnodes', 'n_unodes', 'n_tags']].plot(ax=ax, title='Number of Unitig-Node Tags', lw=4)
    ax.set_xlabel('Read Number')
```
